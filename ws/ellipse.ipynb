## What is an Ellipse?

Given a centering vector, say $c_0$, and a weighting matrix, say $W$, an ellipse can be defined as the points that solve the following inequality:

$$ \{ x : (x - c_0)^\top W (x - c_0) \leq v \} $$

for some scalar $v \in \mathbb{R}$.

See the examples below on how the centering vector and the weighting matrix affect the ellipse.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def quadratic_form(theta, center, weight):
    result = (theta - center).T @ weight @ (theta - center)
    return result
```

```python
def plot_ellipse(center, weight):
    thetas = np.random.uniform(low=-2, high=2, size=(100_000, 2))

    qf = np.array([
        quadratic_form(theta, center, weight) for theta in thetas
    ])

    ellipse = thetas[qf < 0.05]

    plt.scatter(ellipse[:, 0], ellipse[:, 1])
    plt.xlim((-1.5, 1.5))
    plt.ylim((-1.5, 1.5))
```

```python
plot_ellipse(
    center=np.array([0, 0]),  # centered at zero vector
    weight=0.08 * np.eye(2)  # np.eye is the identity matrix
)
```

```python
plot_ellipse(
    center=np.array([1, 0]),
    weight=0.08 * np.eye(2)
)
```

```python
plot_ellipse(
    center=np.array([0, 1]),
    weight=0.08 * np.eye(2)
)
```

```python
plot_ellipse(
    center=np.array([0, 0]),
    weight=0.1 * np.array([[1, 0.5], [0.5, 1]])  # "correlation", i.e. nonzero elements on off-diagonal
)
```

```python
plot_ellipse(
    center=np.array([0, 0]),
    weight=0.1 * np.array([[0.5, 0.5], [0.5, 1]])  # "correlation", i.e. nonzero elements on off-diagonal
)
```
